## Problem Definition

### Project Overview

La rotación de empleados (también conocida como "employee churn") es un problema costoso para las empresas. El coste real de la sustitución de un empleado puede ser bastante elevado. Un estudio realizado por el [Center for American Progress](https://www.americanprogress.org/wp-content/uploads/2012/11/CostofTurnover.pdf) reveló que las empresas suelen pagar alrededor de una quinta parte del salario de un empleado para reemplazarlo, y el coste puede aumentar significativamente si hay que reemplazar a los ejecutivos o a los empleados mejor pagados. En otras palabras, el coste de sustituir a los empleados para la mayoría de las empresas sigue siendo significativo. Esto se debe a la cantidad de tiempo que se invierte en entrevistar y encontrar un sustituto, a las primas de incorporación y a la pérdida de productividad durante varios meses mientras el nuevo empleado se acostumbra a su nueva función. <br>

### Problem Statement

Entender por qué y cuándo es más probable que los empleados se vayan puede conducir a acciones para mejorar la retención de los empleados, así como a la posible planificación de nuevas contrataciones por adelantado. Utilizaré un enfoque sistemático paso a paso utilizando un método que podría ser utilizado para una variedad de problemas de ML. Este proyecto entraría dentro de lo que comúnmente se conoce como "**HR Analytics**", "**People Analytics**". <br>

En este estudio, intentaremos resolver el siguiente planteamiento del problema es: <br>
> ** ¿Cuál es la probabilidad de que un empleado activo abandone la empresa? <br>
¿Cuáles son los principales indicadores de que un empleado abandona la empresa? <br>
¿Qué políticas o estrategias pueden adoptarse a partir de los resultados para mejorar la retención de los empleados? **

Dado que disponemos de datos sobre antiguos empleados, se trata de un problema estándar de **clasificación supervisada** en el que la etiqueta es una variable binaria, 0 (empleado activo), 1 (antiguo empleado). En este estudio, nuestra variable objetivo Y es la probabilidad de que un empleado deje la empresa. <br>

### Feature distribution by target attribute

#### Age

> Las distribuciones de edad de los empleados activos y ex empleados sólo difieren en un año. <br>
La edad media de los ex empleados es de **33,6** años, mientras que **37,6** es la edad media de los empleados actuales.

```python
(mu, sigma) = norm.fit(df_HR.loc[df_HR['Attrition'] == 'Yes', 'Age'])
print(
    'Ex-empleados: edad media = {:.1f} años y desviación estándar = {:.1f}'.format(mu, sigma))
(mu, sigma) = norm.fit(df_HR.loc[df_HR['Attrition'] == 'No', 'Age'])
print('Empleados actuales: edad media = {:.1f} años y desviación estándar = {:.1f}'.format(
    mu, sigma))
```

> Creamos un gráfico de estimación de la densidad del núcleo (KDE) coloreado por el valor del objetivo. Una estimación de la densidad del núcleo (KDE) es una forma no paramétrica de estimar la función de densidad de probabilidad de una variable aleatoria. Nos permitirá identificar si existe una correlación entre la edad del cliente y su capacidad para devolverlo.

```python
plt.figure(figsize=(15,6))
plt.style.use('seaborn-colorblind')
plt.grid(True, alpha=0.5)
sns.kdeplot(df_HR.loc[df_HR['Attrition'] == 'No', 'Age'], label = 'Empleados activos')
sns.kdeplot(df_HR.loc[df_HR['Attrition'] == 'Yes', 'Age'], label = 'Ex-empleados')
plt.xlim(left=18, right=60)
plt.xlabel('Edad (años)')
plt.ylabel('Densidad')
plt.title('Edad vs salidas');
```

## Pre-processing Pipeline

En esta sección, llevamos a cabo los pasos de preprocesamiento de datos para preparar los conjuntos de datos para la implementación del algoritmo de aprendizaje automático.

### Encoding

> Los algoritmos de aprendizaje automático normalmente sólo pueden tener valores numéricos como variables de predicción. Por lo tanto, la codificación de etiquetas es necesaria, ya que codifican las etiquetas categóricas con valores numéricos. Para evitar la introducción de la importancia de las características categóricas con un gran número de valores únicos, utilizaremos tanto la codificación de etiquetas como la codificación de un solo valor, como se muestra a continuación.

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# Create a label encoder object
le = LabelEncoder()
```

```python
print(df_HR.shape)
df_HR.head()
```

```python
# La codificación de etiquetas se utilizará para las columnas con 2 o menos valores únicos
le_count = 0
for col in df_HR.columns[1:]:
    if df_HR[col].dtype == 'object':
        if len(list(df_HR[col].unique())) <= 2:
            le.fit(df_HR[col])
            df_HR[col] = le.transform(df_HR[col])
            le_count += 1
print('{} columns were label encoded.'.format(le_count))
```

```python
# convertir el resto de la variable categórica en dummy
df_HR = pd.get_dummies(df_HR, drop_first=True)
```

> The resulting dataframe has **49 columns** for 1,470 employees.

```python
print(df_HR.shape)
df_HR.head()
```

### Feature Scaling

> El escalado de características utilizando MinMaxScaler esencialmente reduce el rango de tal manera que el rango está ahora entre 0 y n. Los algoritmos de aprendizaje automático funcionan mejor cuando las variables numéricas de entrada caen dentro de una escala similar. En este caso, estamos escalando entre 0 y 5.

```python
# import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 5))
HR_col = list(df_HR.columns)
HR_col.remove('Attrition')
for col in HR_col:
    df_HR[col] = df_HR[col].astype(float)
    df_HR[[col]] = scaler.fit_transform(df_HR[[col]])
df_HR['Attrition'] = pd.to_numeric(df_HR['Attrition'], downcast='float')
df_HR.head()
```

### Splitting data into training and testing sets

## Building Machine Learning Models

### Baseline Algorithms

> Utilicemos primero una serie de algoritmos de base (utilizando hiperparámetros de fábrica) antes de pasar a soluciones más sofisticadas. Los algoritmos considerados en esta sección son: **Regresión logística**, **Bosque aleatorio**, **SVM**, **KNN**, **Clasificador de árbol de decisión**, **NB gaussiano**.

```python
# selección de algoritmos a tener en cuenta y establecimiento de medidas de rendimiento
models = []
models.append(('Logistic Regression', LogisticRegression(solver='liblinear', random_state=7,
                                                         class_weight='balanced')))
models.append(('Random Forest', RandomForestClassifier(
    n_estimators=100, random_state=7)))
models.append(('SVM', SVC(gamma='auto', random_state=7)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('Decision Tree Classifier',
               DecisionTreeClassifier(random_state=7)))
models.append(('Gaussian NB', GaussianNB())
models.append(('XGboost', XGBClassifier(objective='binary:logistic', random_state=33, n_jobs=-1))))
```

> Evaluemos cada modelo por separado y proporcionemos las puntuaciones de precisión y desviación estándar

```python
acc_results = []
auc_results = []
names = []
# set table to table to populate with performance results
col = ['Algorithm', 'ROC AUC Mean', 'ROC AUC STD', 
       'Accuracy Mean', 'Accuracy STD']
df_results = pd.DataFrame(columns=col)
i = 0
# evaluate each model using cross-validation
for name, model in models:
    kfold = model_selection.KFold(
        n_splits=10, random_state=None)  # 10-fold cross-validation

    cv_acc_results = model_selection.cross_val_score(  # accuracy scoring
        model, X_train, y_train, cv=kfold, scoring='accuracy')

    cv_auc_results = model_selection.cross_val_score(  # roc_auc scoring
        model, X_train, y_train, cv=kfold, scoring='roc_auc')

    acc_results.append(cv_acc_results)
    auc_results.append(cv_auc_results)
    names.append(name)
    df_results.loc[i] = [name,
                         round(cv_auc_results.mean()*100, 2),
                         round(cv_auc_results.std()*100, 2),
                         round(cv_acc_results.mean()*100, 2),
                         round(cv_acc_results.std()*100, 2)
                         ]
    i += 1
df_results.sort_values(by=['ROC AUC Mean'], ascending=False)
```

> **Exactitud de la clasificación** es el número de predicciones correctas realizadas como proporción de todas las predicciones realizadas. <br> 
Es la métrica de evaluación más común para los problemas de clasificación. Sin embargo, a menudo se **mal utiliza** ya que sólo es realmente adecuado cuando hay un **número igual de observaciones en cada clase** y todas las predicciones y errores de predicción son igualmente importantes. No es el caso de este proyecto, por lo que una métrica de puntuación diferente puede ser más adecuada.

```python
fig = plt.figure(figsize=(15, 7))
fig.suptitle('Algorithm Accuracy Comparison')
ax = fig.add_subplot(111)
plt.boxplot(acc_results)
ax.set_xticklabels(names)
plt.show()
```

> **El área bajo la curva ROC** (o AUC para abreviar) es una métrica de rendimiento para problemas de clasificación binaria. <br>
El AUC representa la capacidad de un **modelo para discriminar entre clases positivas y negativas**. Un área de 1,0 representa un modelo que hizo todas las predicciones perfectamente. Un área de 0,5 representa un modelo tan bueno como aleatorio.

```python
fig = plt.figure(figsize=(15, 7))
fig.suptitle('Algorithm ROC AUC Comparison')
ax = fig.add_subplot(111)
plt.boxplot(auc_results)
ax.set_xticklabels(names)
plt.show()
```

> Basándonos en nuestro análisis de comparación ROC AUC, **Regresión Logística** y **Random Forest** muestran las puntuaciones medias AUC más altas. Seleccionaremos estos dos algoritmos para un análisis más detallado. Para más detalles sobre estos dos algoritmos, véase más abajo.

**La regresión logística** es un algoritmo de clasificación de aprendizaje automático que se utiliza para predecir la probabilidad de una variable dependiente categórica. La Regresión Logística es un algoritmo de clasificación que no es tan sofisticado como los métodos de conjunto o el método de los árboles de decisión reforzados que se comentan a continuación. Por lo tanto, nos proporciona un buen punto de referencia. 

![title](https://cdn-images-1.medium.com/max/1600/0*vRhSdZ_k4wrP6Bl8.jpg)

**Random Forest** es un método de aprendizaje automático popular y versátil que es capaz de resolver tanto la regresión como la clasificación. Random Forest es una marca de aprendizaje de conjunto, ya que se basa en un conjunto de árboles de decisión. Agrega árboles de clasificación (o de regresión). Un árbol de decisión se compone de una serie de decisiones que pueden utilizarse para clasificar una observación en un conjunto de datos.

Random Forest ajusta una serie de clasificadores de árboles de decisión en varias **submuestras del conjunto de datos** y utiliza el **promedio** para mejorar la precisión predictiva y controlar el sobreajuste. Random Forest puede manejar un gran número de características, y es útil para estimar cuáles de sus variables son importantes en los datos subyacentes que se están modelando.

### Logistic Regression

> Veamos con más detalle el uso del algoritmo de Regresión Logística. Utilizaré la validación cruzada de 10 veces para entrenar nuestro modelo de regresión logística y estimar su puntuación AUC.

```python
kfold = model_selection.KFold(n_splits=10, random_state=None)
modelCV = LogisticRegression(solver='liblinear',
                             class_weight="balanced", 
                             random_state=None)
scoring = 'roc_auc'
results = model_selection.cross_val_score(
    modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("AUC score (STD): %.2f (%.2f)" % (results.mean(), results.std()))
```

#### Fine-tuning

> GridSearchCV permite el ajuste fino de los hiperparámetros mediante la búsqueda de los valores de los parámetros especificados para un estimador.

```python
param_grid = {'C': np.arange(1e-03, 2, 0.01)} # hyper-parameter list to fine-tune
log_gs = GridSearchCV(LogisticRegression(solver='liblinear', # setting GridSearchCV
                                         class_weight="balanced", 
                                         random_state=None), return_train_score=True, 
                      param_grid=param_grid,
                      scoring='roc_auc',
                      cv=10)

log_grid = log_gs.fit(X_train, y_train)
log_opt = log_grid.best_estimator_
results = log_gs.cv_results_

print('='*20)
print("best params: " + str(log_gs.best_estimator_))
print("best params: " + str(log_gs.best_params_))
print('best score:', log_gs.best_score_)
print('='*20)
```

> Como se muestra arriba, los resultados de GridSearchCV nos proporcionaron un hiperparámetro ajustado utilizando ROC_AUC como métrica de puntuación.

#### Evaluation

```python
## Confusion Matrix
cnf_matrix = metrics.confusion_matrix(y_test, log_opt.predict(X_test))
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print('Precisión del clasificador de regresión logística en el conjunto de pruebas: {:.2f}'.format(log_opt.score(X_test, y_test)*100))
```

> La matriz de confusión nos proporciona una representación mucho más detallada de la puntuación de precisión y de lo que ocurre con nuestras etiquetas: sabemos exactamente qué etiquetas se han predicho correctamente y cuáles no.

```python
# Informe de clasificación de la regresión logarítmica optimizada
log_opt.fit(X_train, y_train)
print(classification_report(y_test, log_opt.predict(X_test)))
```

> La puntuación AUC resultante es mayor que la mejor puntuación durante el paso de optimización. La predicción de las probabilidades de una etiqueta concreta nos proporciona una medida de la probabilidad de que un empleado abandone la empresa.

```python
log_opt.fit(X_train, y_train) # ajustar el modelo optimizado a los datos de entrenamiento
probs = log_opt.predict_proba(X_test) # predict probabilities
probs = probs[:, 1] # sólo mantendremos las probabilidades asociadas a la salida del empleado
logit_roc_auc = roc_auc_score(y_test, probs) # calcular la puntuación AUC utilizando el conjunto de datos de prueba
print('AUC score: %.3f' % logit_roc_auc)
```

La línea verde representa la curva ROC de un clasificador puramente aleatorio; un buen clasificador se mantiene lo más alejado posible de esa línea (hacia la esquina superior izquierda). <br>

A medida que la empresa genera más datos sobre sus empleados (sobre las nuevas incorporaciones y las bajas recientes), el algoritmo puede volver a entrenarse utilizando los datos adicionales y, en teoría, generar predicciones más precisas para identificar a los **empleados de alto riesgo** de abandonar la empresa, basándose en la etiqueta probabilística asignada por el algoritmo a cada variable de características (es decir, al empleado).

A los empleados se les puede asignar una "Categoría de riesgo" basada en la etiqueta prevista, de forma que:
- **Low-risk** for employees with label < 0.6
- **Medium-risk** for employees with label between 0.6 and 0.8
- **High-risk** for employees with label > 0.8 <br>

### Strategic Retention Plan

- Los indicadores más fuertes de la gente que deja la empresa son:
    - **Ingresos mensuales**: las personas con salarios más altos tienen menos probabilidades de abandonar la empresa. Por lo tanto, hay que esforzarse por reunir información sobre los puntos de referencia del sector en el mercado local actual para determinar si la empresa ofrece salarios competitivos.
    - **Horas extras**: las personas que hacen horas extras son más propensas a abandonar la empresa. Por lo tanto, hay que esforzarse en dimensionar adecuadamente los proyectos por adelantado con el apoyo y la mano de obra adecuados para reducir el uso de las horas extraordinarias.
    - **Años con el gerente**: Un gran número de personas que abandonan la empresa lo hacen 6 meses después de que lo hagan sus jefes actuales. Utilizando los detalles del Jefe de Línea para cada empleado, se puede determinar qué Jefe ha experimentado el mayor número de empleados que han renunciado en el último año. Aquí se pueden utilizar varias métricas para determinar si se deben tomar medidas con un Jefe de línea: 
        - el número de empleados que dependen de los directivos y que presentan un alto índice de rotación: esto indicaría que puede ser necesario revisar la estructura de la organización para mejorar la eficiencia
        - número de años que el jefe de línea lleva en un determinado puesto: esto puede indicar que los empleados pueden necesitar formación en gestión o que se les asigne un mentor (idealmente un ejecutivo) en la organización
        - patrones de los empleados que han renunciado: esto puede indicar patrones recurrentes en los empleados que se van, en cuyo caso se pueden tomar medidas en consecuencia.
    - **Edad**: Los empleados relativamente jóvenes, entre 25 y 35 años, son más propensos a marcharse. Por lo tanto, hay que esforzarse por articular claramente la visión a largo plazo de la empresa y el encaje de los empleados jóvenes en esa visión, así como ofrecer incentivos en forma de vías claras de promoción, por ejemplo.
    - **Distancia a casa**: Los empleados que viven más lejos de su casa tienen más probabilidades de abandonar la empresa. Por lo tanto, hay que esforzarse por ofrecer apoyo en forma de transporte de la empresa para los grupos de empleados que se marchan de la misma zona, o en forma de subsidio de transporte. Probablemente no se recomiende la selección inicial de los empleados en función de su lugar de residencia, ya que se consideraría una forma de discriminación, siempre y cuando los empleados lleguen a tiempo al trabajo todos los días.
    - **Años de trabajo**: Los empleados con más experiencia tienen menos probabilidades de marcharse. Los empleados que tienen entre 5 y 8 años de experiencia deben ser identificados como potencialmente con mayor riesgo de abandono.
    - **Años en la Empresa**: Los empleados fieles a la empresa son menos propensos a marcharse. Los empleados que cumplen dos años deben ser identificados como potencialmente con mayor riesgo de irse.

Debe elaborarse un **"Plan de retención "** estratégico para cada grupo de **categorías de riesgo**. Además de los pasos sugeridos para cada característica enumerada anteriormente, pueden iniciarse reuniones cara a cara entre un representante de RRHH y los empleados para los **empleados de riesgo medio** y los **empleados de riesgo alto** para discutir las condiciones de trabajo. Asimismo, una reunión con el jefe de línea de esos empleados permitiría discutir el ambiente de trabajo dentro del equipo y si se pueden tomar medidas para mejorarlo.
